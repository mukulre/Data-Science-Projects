```python
import pandas as pd
import numpy as np

# Load the dataset
dataset = pd.read_csv("/content/Ads_CTR_Optimisation (1).csv")  # Ensure the filename matches your download

# Parameters
n_rounds = dataset.shape[0] # Use the number of rows for the number of rounds
n_ads = dataset.shape[1] # Use the number of columns for the number of ads
epsilon = 0.1

# Tracking variables
numbers_of_selections = [0] * n_ads
sums_of_rewards = [0] * n_ads
ads_selected = []
total_reward = 0

# Epsilon-Greedy Loop
for n in range(n_rounds):
    if np.random.rand() < epsilon:
        ad = np.random.randint(n_ads)  # Explore random ad
    else:
        average_rewards = [sums_of_rewards[i] / numbers_of_selections[i] if numbers_of_selections[i]>0 else 0 for i in range(n_ads)]
        ad = np.argmax(average_rewards)  # Exploit best ad
    reward = dataset.values[n, ad]
    numbers_of_selections[ad] += 1
    sums_of_rewards[ad] += reward
    ads_selected.append(ad)
    total_reward += reward

# Results
best_ad = np.argmax(sums_of_rewards)
print(f"Best ad index: {best_ad}")
print(f"Total reward: {total_reward}")

# Optional: Summary DataFrame
summary = pd.DataFrame({
    'Ad': range(n_ads),
    'Selections': numbers_of_selections,
    'Reward': sums_of_rewards
})
print(summary.sort_values('Reward', ascending=False).head(5))  # Top 5 ads
```
