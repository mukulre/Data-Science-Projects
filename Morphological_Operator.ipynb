## Morphological Operators in Image Processing

Morphological operators are image processing techniques based on **shapes**.  
They are mostly used on **binary images** (black & white).

### Why use morphological operations?
- To remove noise
- To enhance shapes
- To separate objects
- To fill gaps or holes
- To extract boundaries

### Common Operators:
1. **Erosion** – Shrinks white regions.
2. **Dilation** – Expands white regions.
3. **Opening** – Erosion followed by Dilation (removes small noise).
4. **Closing** – Dilation followed by Erosion (fills small holes).
5. **Morphological Gradient** – Boundary extraction.
6. **Top-hat** – Extracts small bright regions.
7. **Black-hat** – Extracts small dark regions on bright background.


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def load_img():
    img_canvas = np.zeros((600,600))
    font=cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img_canvas,'ABCDE',(50,300),font,5,(255,255,255),25)
    return img_canvas
```

```python
def disp_img(img):
    fig=plt.figure(figsize=(10,10))
    plt.imshow(img,cmap='gray')
```

```python
img1 = load_img()
```

```python
disp_img(img1)
```

```python
kernel = np.ones((9,9),np.uint8) #square structuring element of size 5x5
```

```python
#morphological operations - erode
img2 = cv2.erode(img1,kernel,iterations=1)

fig=plt.figure(figsize=(15,15))
fig.add_subplot(121)
plt.imshow(img1,cmap='gray')
fig.add_subplot(122)
plt.imshow(img2,cmap='gray')

#if you see the output images, in the eroded image the boundaries of the text are eroded
```

```python
#morphological operations - dilate
img2 = cv2.dilate(img1,kernel,iterations=1)

fig=plt.figure(figsize=(5,5))
fig.add_subplot(121)
plt.imshow(img1,cmap='gray')
fig.add_subplot(122)
plt.imshow(img2,cmap='gray')

#if you see the output images, in the dilated image the boundaries of the text are expanded
```

```python
k2 = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
```

```python
white_noise = np.random.randint(low=0,high=2,size=(600,600))
white_noise = white_noise*255
```

```python
disp_img(white_noise)
```

```python
noisy_img1= white_noise + img1
disp_img(noisy_img1)
```

```python
#morphological operations - opening
img_open = cv2.morphologyEx(noisy_img1,cv2.MORPH_OPEN,kernel,iterations=1)
disp_img(img_open)
```

```python
#morphological operations - closing
img_close = cv2.morphologyEx(noisy_img1,cv2.MORPH_CLOSE,kernel,iterations=1)
disp_img(img_close)
```

```python
#morphological operations - closing
img_gradient = cv2.morphologyEx(noisy_img1,cv2.MORPH_GRADIENT,kernel,iterations=1)
disp_img(img_gradient)
```
